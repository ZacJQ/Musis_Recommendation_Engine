```python
import ast

metadata_df3_copy = metadata_df3
metadata_df3_copy['artist_name'] = metadata_df3['artist_name'].apply(lambda x: x.split('[')[1].split(']')[0].strip("'") if isinstance(x, str) and '[' in x and ']' in x else x)
```

```python
# Remove leading and trailing single quotes from each term in the 'artist_name' column
def strip_quotes(string):
    terms = [term.strip("'") for term in string.split(",")]
    terms = [term.strip() for term in terms]  # Remove any extra whitespace
    test = [x.strip("'") for x in terms]
    return ', '.join(test)

metadata_df3_copy['artist_name'] = metadata_df3_copy['artist_name'].apply(strip_quotes)
```
